```python
ped_file = '../data/v34.vcf.ped'
identicals_file = '../sibpair_similarity/ihart_identicals.txt'
#ped_file = '../data/ssc.ped'
#ped_file = '../data/spark.ped'

families = dict()
child_id_to_affected = dict()
child_id_to_sex = dict()

with open(ped_file, 'r') as f:
    for line in f:
        pieces = line.strip().split('\t')
        
        if len(pieces) >= 6:
            fam_id, child_id, f_id, m_id, sex, disease_status = pieces[0:6]

            if m_id != '0' and f_id != '0':
                if (fam_id, m_id, f_id) not in families:
                    families[(fam_id, m_id, f_id)] = [m_id, f_id]
                families[(fam_id, m_id, f_id)].append(child_id)
            
                child_id_to_affected[child_id] = disease_status
                child_id_to_sex[child_id] = sex
```

```python
# pull twins
identicals = set()
with open(identicals_file, 'r') as f:
	for line in f:
		pieces = line.strip().split('\t')
		identicals.update(pieces)
print('identicals', len(identicals))
```

```python
import random

with open(ped_file + '.famsize.ped', 'w+') as f:
    for famkey, inds in families.items():
        if len(inds) == 6 and len([x for x in inds if x in identicals]) == 0:
            print(famkey, inds)
            children = inds[2:]
            chosen = []
            for i in range(len(children)):
                choice = random.choice(children)
                chosen.append(choice)
                children.remove(choice)
                
                new_famkey = '%s.%d' % ('.'.join(famkey), i+3)
                for child in chosen:
                    f.write('\t'.join([new_famkey, child, inds[1], inds[0], child_id_to_sex[child], child_id_to_affected[child]]) + '\n')

                
```
