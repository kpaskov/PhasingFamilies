```python
import json
from collections import defaultdict
```

```python
in_ped = set()
with open('../../DATA/spark/sparkfam.ped', 'r') as f:
    for line in f:
        pieces = line.strip().split('\t')
        in_ped.add(pieces[1])
        in_ped.add(pieces[2])
        in_ped.add(pieces[3])
```

```python
with open('../../DATA/spark/genotypes/samples.json', 'r') as f:
    spark_samples = json.load(f)
```

```python
with open('../../DATA/spark/genotypes/wes1_array/samples.json', 'r') as f:
    wes1_spark_samples = json.load(f)
    
with open('../../DATA/spark/genotypes/wes2_array/samples.json', 'r') as f:
    wes2_spark_samples = json.load(f)
    
with open('../../DATA/spark/genotypes/wes3_array/samples.json', 'r') as f:
    wes3_spark_samples = json.load(f)
    
with open('../../DATA/spark/genotypes/wgs1_B01_array/samples.json', 'r') as f:
    wgs1b01_spark_samples = json.load(f)
    
with open('../../DATA/spark/genotypes/wgs1_B02_array/samples.json', 'r') as f:
    wgs1b02_spark_samples = json.load(f)
    
with open('../../DATA/spark/genotypes/wgs2_array/samples.json', 'r') as f:
    wgs2_spark_samples = json.load(f)
    
with open('../../DATA/spark/genotypes/wgs3_array/samples.json', 'r') as f:
    wgs3_spark_samples = json.load(f)
```

```python
print(spark_samples[:2])
print(wes1_spark_samples[:2], wes2_spark_samples[:2], wes3_spark_samples[:2])
print(wgs1b01_spark_samples[:2], wgs1b02_spark_samples[:2], wgs2_spark_samples[:2], wgs3_spark_samples[:2])
```

```python
print('orig', len(set(spark_samples) & in_ped)/len(spark_samples), len(set(spark_samples) - in_ped))
print('wes1', len(set(wes1_spark_samples) & in_ped)/len(wes1_spark_samples), len(set(wes1_spark_samples) - in_ped))
print('wes2', len(set(wes2_spark_samples) & in_ped)/len(wes2_spark_samples), len(set(wes2_spark_samples) - in_ped))
print('wes3', len(set(wes3_spark_samples) & in_ped)/len(wes3_spark_samples), len(set(wes3_spark_samples) - in_ped))
print('wgs1b01', len(set(wgs1b01_spark_samples) & in_ped)/len(wgs1b01_spark_samples), len(set(wgs1b01_spark_samples) - in_ped))
print('wgs1b02', len(set(wgs1b02_spark_samples) & in_ped)/len(wgs1b02_spark_samples), len(set(wgs1b02_spark_samples) - in_ped))
print('wgs2', len(set(wgs2_spark_samples) & in_ped)/len(wgs2_spark_samples), len(set(wgs2_spark_samples) - in_ped))
print('wgs3', len(set(wgs3_spark_samples) & in_ped)/len(wgs3_spark_samples), len(set(wgs3_spark_samples) - in_ped))
```

```python
print('orig', len(set(spark_samples) - set(spark_samples)))
print('wes1', len(set(wes1_spark_samples) - set(spark_samples)))
print('wes2', len(set(wes2_spark_samples) - set(spark_samples)))
print('wes3', len(set(wes3_spark_samples) - set(spark_samples)))
print('wgs1b01', len(set(wgs1b01_spark_samples) - set(spark_samples)))
print('wgs1b02', len(set(wgs1b02_spark_samples) - set(spark_samples)))
print('wgs2', len(set(wgs2_spark_samples) - set(spark_samples)))
print('wgs3', len(set(wgs3_spark_samples) - set(spark_samples)))
```

```python
print('orig', len(set(spark_samples) & set(spark_samples)))
print('wes1', len(set(wes1_spark_samples) & set(spark_samples)))
print('wes2', len(set(wes2_spark_samples) & set(spark_samples)))
print('wes3', len(set(wes3_spark_samples) & set(spark_samples)))
print('wgs1b01', len(set(wgs1b01_spark_samples) & set(spark_samples)))
print('wgs1b02', len(set(wgs1b02_spark_samples) & set(spark_samples)))
print('wgs2', len(set(wgs2_spark_samples) & set(spark_samples)))
print('wgs3', len(set(wgs3_spark_samples) & set(spark_samples)))
```

```python
sample_to_datasets = defaultdict(list)

for sample in wes1_spark_samples:
    sample_to_datasets[sample].append('wes1')
for sample in wes2_spark_samples:
    sample_to_datasets[sample].append('wes2')
for sample in wes3_spark_samples:
    sample_to_datasets[sample].append('wes3')
for sample in wgs1b01_spark_samples:
    sample_to_datasets[sample].append('wgs1_b01')
for sample in wgs1b02_spark_samples:
    sample_to_datasets[sample].append('wgs1_b02')
for sample in wgs2_spark_samples:
    sample_to_datasets[sample].append('wgs2')
for sample in wgs3_spark_samples:
    sample_to_datasets[sample].append('wgs3')
    
for sample, datasets in sample_to_datasets.items():
    if len(datasets)>1:
        print(sample, datasets)
```
