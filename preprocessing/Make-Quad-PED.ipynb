```python
#ped_file = '../data/160826.ped'
#ped_file = '../data/ssc.ped'
ped_file = '../data/spark.ped'

families = dict()
child_id_to_affected = dict()
child_id_to_sex = dict()

with open(ped_file, 'r') as f:
    for line in f:
        pieces = line.strip().split('\t')
        
        if len(pieces) >= 6:
            fam_id, child_id, f_id, m_id, sex, disease_status = pieces[0:6]

            if 'LCL' not in child_id and m_id != '0' and f_id != '0':
                if (fam_id, m_id, f_id) not in families:
                    families[(fam_id, m_id, f_id)] = [m_id, f_id]
                families[(fam_id, m_id, f_id)].append(child_id)
            
                child_id_to_affected[child_id] = disease_status
                child_id_to_sex[child_id] = sex
```

```python
from itertools import combinations

with open(ped_file + '.quads.ped', 'w+') as f:
    for famkey, inds in families.items():
        print(famkey, inds)
        for i, j in combinations(inds[2:], 2):
            # create new quad family with these two children
            quad_famkey = '%s.%s.%s' % (famkey[0], i, j)
            f.write('\t'.join([quad_famkey, i, famkey[2], famkey[1], child_id_to_sex[i], child_id_to_affected[i]]) + '\n')
            f.write('\t'.join([quad_famkey, j, famkey[2], famkey[1], child_id_to_sex[j], child_id_to_affected[j]]) + '\n')
```

# Now make LCL ped

```python
ped_file = '../data/160826.ped'

# pull LCL families
families = dict()
child_id_to_affected = dict()
child_id_to_sex = dict()

with open(ped_file, 'r') as f:
    for line in f:
        pieces = line.strip().split('\t')
        
        if len(pieces) >= 6:
            fam_id, child_id, f_id, m_id, sex, disease_status = pieces[0:6]
            
            if 'LCL' in child_id:
                families[(fam_id, m_id, f_id)] = [m_id, f_id, child_id]
                
                child_id_to_affected[child_id] = disease_status
                child_id_to_sex[child_id] = sex
            
with open(ped_file, 'r') as f:
    for line in f:
        pieces = line.strip().split('\t')
        
        if len(pieces) >= 6:
            fam_id, child_id, f_id, m_id, sex, disease_status = pieces[0:6]
            
            if (fam_id, m_id, f_id) in families and m_id != '0' and f_id != '0' \
                and child_id != families[(fam_id, m_id, f_id)][2] \
                and child_id != families[(fam_id, m_id, f_id)][2][:-4]:
                families[(fam_id, m_id, f_id)].append(child_id)
            
                child_id_to_affected[child_id] = disease_status
                child_id_to_sex[child_id] = sex
```
