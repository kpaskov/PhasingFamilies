```python
old_directory = '../phased_ihart_quad'
new_directory = '../phased_ihart_quad_broader_unknowns'
```

```python
from os import listdir

phased_files = []
for filename in listdir(old_directory):
    if filename.endswith('phased.txt'):
        phased_files.append(filename)
print('Files', len(phased_files))
```

```python
for filename in listdir(old_directory):
    if filename.endswith('families.txt'):
        with open('%s/%s' % (old_directory, filename), 'r') as f, \
        open('%s/%s' % (new_directory, filename), 'w+') as nf:
            for line in f:
                nf.write(line)
```

```python
from itertools import islice

for filename in phased_files:
    print(filename)
    # read in data
    lines = []
    with open('%s/%s' % (old_directory, filename), 'r') as f:
        header = next(f) # skip header

        for line in f:
            pieces = line.strip().split('\t')
            pieces = [pieces[0]] + [int(x) for x in pieces[1:]]
            lines.append(pieces)

    buffer_size = 1000000

    # extend mask forwards
    final_lines = [lines[0]]
    extended = 0
    buffer = 0

    for line in lines[1:]:
        if final_lines[-1][0] == line[0] and final_lines[-1][-7] == 1 and line[-7] == 0:
            buffer = buffer_size
            extended += 1

        if buffer > 0:
            if line[-2] < buffer:
                final_lines.append(line[:-7] + [1] + line[-6:])
                buffer -= line[-2]
            else:
                final_lines.append(line[:-7] + [1] + [line[-6], line[-6]+buffer-1, -1, -1, buffer, -1])
                final_lines.append(line[:-6] + [line[-6]+buffer, line[-5], -1, -1, line[-2]-buffer, -1])                 
                buffer = 0 
        else:
            final_lines.append(list(line))
    print('Extended forwards', extended)

    # extend mask backwards
    lines = list(reversed(final_lines))

    final_lines = [lines[0]]
    extended = 0
    buffer = 0

    for line in lines[1:]:
        if final_lines[-1][0] == line[0] and final_lines[-1][-7] == 1 and line[-7] == 0:
            buffer = buffer_size
            extended += 1

        if buffer > 0:
            if line[-2] < buffer:
                final_lines.append(line[:-7] + [1] + line[-6:])
                buffer -= line[-2]
            else:
                final_lines.append(line[:-7] + [1] + [line[-5]-buffer, line[-5], -1, -1, buffer, -1])
                final_lines.append(line[:-6] + [line[-6], line[-5]-buffer-1, -1, -1, line[-2]-buffer, -1])                 
                buffer = 0 
        else:
            final_lines.append(list(line))
    print('Extended backwards', extended)


    final_lines = reversed(final_lines)
    
    with open('%s/%s' % (new_directory, filename), 'w+') as f:
        f.write(header)
        for line in final_lines:
            f.write('\t'.join([str(x) for x in line]) + '\n')
```
